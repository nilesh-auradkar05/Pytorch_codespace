## Random Tensors

Why Random Tensors?

Many NN start with tensors full of random numbers and then adjust those random numbers to better represent the data.

1. start with random number tensors.
2. look at data.
3. update random numbers to some calculated value.
4. repeat steps 2 and 3.

```python
# Random tensors of shape (3, 4)(rows, columns)
RANDOM_TENSOR = torch.rand(3, 4)
print(RANDOM_TENSOR)
print(f"dims of RANDOM_TENSOR: {RANDOM_TENSOR.ndim}")
print(f"shape of RANDOM_TENSOR: {RANDOM_TENSOR.shape}")
```

## Matrix Multiplication or Dot product


 

There are two main rules to perform dot product on matrix/tensor:
1. The **inner dimensions** should match:
* `(3, 2) * (3, 2)` :- would cause error.
* `(3, 2) * (2, 3)` :- will produce a result.

2. The resulting matrix has the shape of **outer dimensions**:
* `(2, 3) * (3, 2)` => `(2, 2)`
* `(3, 2) * (2, 3)` => `(3, 3)`

```python
# Matrix Multiplication
TENSOR_1 = torch.randint(1, 11, size=(3, 2),
                         dtype=torch.bfloat16,
                         device="cuda",
                         requires_grad=False)

TENSOR_2 = torch.rand(size=(3, 2),
                      dtype=torch.bfloat16,
                      device="cuda",
                      requires_grad=False)
print(TENSOR_1)
print(TENSOR_2)
```

```python
print(TENSOR_1, "*", TENSOR_2)
print(f"Equals: {TENSOR_1.matmul(TENSOR_2)}")
```

## Method to handle mismatch Shape error in dot product of tensors

* **Using Transpose (tensor.T)**

```python
TENSOR_2 = TENSOR_2.T
```

```python
TENSOR_1.shape, TENSOR_2.shape
```

```python
print(TENSOR_1, "\n", "*", "\n", TENSOR_2)
print(f"Equals: {TENSOR_1.matmul(TENSOR_2)}")
```

## Reshaping, Stacking, Squeezing, Unsqueezing

* Reshaping :- reshapes an input tensor to a defined shape
* View :- return a view of an input tensor of certain shape but keep the same memory as the original tensor
* Stacking :- combine multiple tensors on top of each other (vstack) or side by side (hstack).
* Squeezing :- remove all `1` dimensions from a tensor
* Unsqueezing :- add a `1` dimension to a target tensor.
* Permute :- Return a view of the input with dimensions permuted(swapped) in a certain way

```python
# sample tensor
TENSOR_MANIPULATION = torch.arange(1., 11.,
                                   dtype=torch.bfloat16,
                                   device=None)
TENSOR_MANIPULATION, TENSOR_MANIPULATION.shape
```

```python
# Reshaping tensor
## Note: the new shape must be compatible with the original shape
try:
    TENSOR_MANIPULATION_reshape_1 = TENSOR_MANIPULATION.reshape(1, 9)
except Exception as e:
    print(e)

TENSOR_MANIPULATION_reshape_2 = TENSOR_MANIPULATION.reshape(1, 10)
print(TENSOR_MANIPULATION_reshape_2, TENSOR_MANIPULATION_reshape_2.shape)

TENSOR_MANIPULATION_reshape_3 = TENSOR_MANIPULATION.reshape(5, 2)
print(TENSOR_MANIPULATION_reshape_3, TENSOR_MANIPULATION_reshape_3.shape)
```

```python
# changing view of tensor
TENSOR_MANIPULATION_view = TENSOR_MANIPULATION.view(5, 2)
print(TENSOR_MANIPULATION_view, "\n", TENSOR_MANIPULATION_view.shape)
```

```python
# changes the value of all row in 1st index for both view and original tensor because they share same memory space
TENSOR_MANIPULATION_view[:, 1] = 11.2
print(TENSOR_MANIPULATION_view, "\n", TENSOR_MANIPULATION)
```

```python
TENSOR_MANIPULATION_vstack = torch.vstack((TENSOR_MANIPULATION,
                                           TENSOR_MANIPULATION,
                                           TENSOR_MANIPULATION,
                                           TENSOR_MANIPULATION))
TENSOR_MANIPULATION_vstack
```

```python
TENSOR_MANIPULATION_stack = torch.stack((TENSOR_MANIPULATION,
                                         TENSOR_MANIPULATION,
                                         TENSOR_MANIPULATION,
                                         TENSOR_MANIPULATION),
                                         dim=0)
TENSOR_MANIPULATION_stack
```

```python
TENSOR_MANIPULATION_hstack = torch.hstack((TENSOR_MANIPULATION,
                                           TENSOR_MANIPULATION,
                                           TENSOR_MANIPULATION,
                                           TENSOR_MANIPULATION))
TENSOR_MANIPULATION_hstack
```

```python
TENSOR_MANIPULATION_stack = torch.stack((TENSOR_MANIPULATION,
                                         TENSOR_MANIPULATION,
                                         TENSOR_MANIPULATION,
                                         TENSOR_MANIPULATION),
                                         dim=1)
TENSOR_MANIPULATION_stack
```

```python
TENSOR_MANIPULATION_stack = torch.stack((TENSOR_MANIPULATION,
                                         TENSOR_MANIPULATION,
                                         TENSOR_MANIPULATION,
                                         TENSOR_MANIPULATION),
                                         dim=-1)
TENSOR_MANIPULATION_stack
```

```python
TENSOR_MANIPULATION_stack = torch.stack((TENSOR_MANIPULATION,
                                         TENSOR_MANIPULATION,
                                         TENSOR_MANIPULATION,
                                         TENSOR_MANIPULATION),
                                         dim=-2)
TENSOR_MANIPULATION_stack
```

```python
# torch.squeeze() removes all 1 dimensions from the tensor
# original tensor
print(f"Unsqueezed Tensor: {TENSOR_MANIPULATION.unsqueeze(0)}")
print(f"Unsqueezed Tensor Shape: {TENSOR_MANIPULATION.unsqueeze(0).shape}")

# Remove extra 1D dimensions
TENSOR_MANIPULATION_squeezed = torch.squeeze(TENSOR_MANIPULATION)
print(f"Squeezed Tensor: {TENSOR_MANIPULATION_squeezed}")
print(f"Squeezed Tensor Shape: {TENSOR_MANIPULATION_squeezed.shape}")
```

```python
# torch.permute() rearranges the dimensions of the tensor
ORIGINAL_TENSOR = torch.randn(size=(224, 224, 3))

# permuted tensor
PERMUTED_TENSOR = torch.permute(ORIGINAL_TENSOR, (2, 0, 1))

print(f"Original Tensor Shape: {ORIGINAL_TENSOR.shape}")
# print(f"Permuted Tensor: {PERMUTED_TENSOR}")
print(f"Permuted Tensor Shape: {PERMUTED_TENSOR.shape}")
```

```python
print(ORIGINAL_TENSOR[0, 0, 0])
ORIGINAL_TENSOR[0, 0, 0] = 211

print(ORIGINAL_TENSOR[0, 0, 0])
print(PERMUTED_TENSOR[0, 0, 0])
```
